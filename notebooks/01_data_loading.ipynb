# 01 — Data Loading & Cleaning
Load raw sales data, clean it, run sanity checks, and save the processed file.

```python
import pandas as pd
import matplotlib.pyplot as plt

# ── Load raw data ─────────────────────────────────────────────
df = pd.read_csv('../data/raw/train.csv')
df.head()
```

```python
# ── Clean ─────────────────────────────────────────────────────
df.columns = df.columns.str.lower()
df['date'] = pd.to_datetime(df['date'])
df = df.sort_values(['store', 'item', 'date'])

print('Missing values:')
print(df.isnull().sum())
```

```python
# ── Sanity checks ─────────────────────────────────────────────
print('Rows:  ', len(df))
print('Stores:', df['store'].nunique())
print('Items: ', df['item'].nunique())
print('Dates: ', df['date'].min().date(), '→', df['date'].max().date())
```

```python
# ── Quick visual ──────────────────────────────────────────────
sample = df[df['item'] == 1]

plt.figure(figsize=(10, 4))
plt.plot(sample['date'], sample['sales'])
plt.title('Item 1 Daily Sales (all stores)')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()
```

```python
# ── Save processed ────────────────────────────────────────────
df.to_csv('../data/processed/processed_sales.csv', index=False)
print('Saved to data/processed/processed_sales.csv')
```
